```python
#But here we will see that how we will convert the CSV data inot series 
data =  pd.read_csv('datasets/AirPassengers.csv')
data.head()
```

```python
# Creating a TimeSeries, specifying the time and value columns
series = TimeSeries.from_dataframe(data, "Month", "#Passengers" )
```

# Split Dataset

```python
train, val = series[:-36], series[-36:]
```

# Scaling ?

# Train Model

```python
"""The dataset is split into:

Training Set:   This subset is used to train the model. The model learns the patterns and 
                relationships from this data.

Validation Set:  This subset is used to tune the model's parameters and make decisions about which model to select.
                 It helps in evaluating the model's performance and preventing overfitting.


Test Set:        This subset is used to evaluate the model's performance after it has been trained and validated. 
                    It serves as an unbiased evaluation of the model's generalization ability."""








from darts.models import ExponentialSmoothing

model = ExponentialSmoothing()
model.fit(train)
```

# Evaluation ?

# Forecasting / Predicting

```python
prediction = model.predict(len(val), num_samples=1000)
prediction
```

# Visualize Results with Matplotlib

```python
import matplotlib.pyplot as plt

series.plot()
prediction.plot(label="forecast", low_quantile=0.5, high_quantile=0.95)
plt.legend()
```

# Trying the Different Models

```python
from darts.models import TBATS, AutoARIMA, Theta
from darts.metrics import mape # For Evaluating the matrix, we are importing the mape


#Creating a general function that will Work on Each Model

def eval_model(model, model_name):
    model.fit(train)                                                               #Train
    forecast = model.predict(len(val))                                             #Predict     
    print("model {} obtains MAPE: {:.2f}%".format(model, mape(val, forecast )))   #Evaluation
    train.plot(label="train")                                                     #Visualization   
    val.plot(label="true")                                                                
    forecast.plot(label=model_name)
```

```python
#Using the TBATS Model
eval_model(TBATS(), 'TBATS')
```

```python
#Using the AutoARIMA Model
eval_model(AutoARIMA(), 'AutoARIMA')
```

```python
#Using the Theta Model
eval_model(AutoARIMA(), 'AutoARIMA')
```
